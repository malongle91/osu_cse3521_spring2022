## Cross Validation

#### Another example on the iris dataset here: https://www.ritchieng.com/machine-learning-cross-validation/

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#create DataFrame
np.random.seed(314)
n = 50
x = np.linspace(-2,2, n)
y = 3*x**3 + 5*x**2 + 0.5*x + 20 +np.random.normal(loc = 0, scale = 4, size = n) # a 3 polynomial model


df = pd.DataFrame({'x': x,
                   'x2': x**2, 
                   'x3':x**3,
                   'x4':x**4,
                   'x5':x**5,
                   'x6':x**6,
                   'x7':x**7,
                   'x8':x**8,
                   'x9':x**9,
                   'x10':x**10,
                   'x11':x**11,
                   'x12':x**12,
                   'x13':x**13,
                   'x14':x**14,
                   'x15':x**15,
                   'y': y})

#create scatterplot of x vs. y
plt.scatter(df.x, df.y)
```

```python
df
```

```python
#fit polynomial models up to degree 5
model1 = np.poly1d(np.polyfit(df.x, df.y, 1))
model2 = np.poly1d(np.polyfit(df.x, df.y, 2))
model3 = np.poly1d(np.polyfit(df.x, df.y, 3))
model5 = np.poly1d(np.polyfit(df.x, df.y, 5))
model7 = np.poly1d(np.polyfit(df.x, df.y, 7))
model15 = np.poly1d(np.polyfit(df.x, df.y, 15))

fig=plt.figure(figsize=(12,8), dpi= 100)

#create scatterplot
polyline = np.linspace(-2, 2, n)
plt.scatter(df.x, df.y)

#add fitted polynomial lines to scatterplot 
plt.plot(polyline, model1(polyline), color='green')
plt.plot(polyline, model2(polyline), color='red')
plt.plot(polyline, model3(polyline), color='purple')
plt.plot(polyline, model5(polyline), color='blue')
plt.plot(polyline, model7(polyline), color='orange')
plt.plot(polyline, model15(polyline), color='black')
plt.legend(['Degree 1', 'Degree 2', 'Degree 3', 'Degree 5', 'Degree 7', 'Degree 15'], loc='upper left')


plt.show()
```

```python
import statsmodels.api as sm
from statsmodels.tools.eval_measures import rmse


model_list = [sm.OLS(df.y, sm.add_constant(df.iloc[:,0:x])) for x in range(16)]
model_fits = [x.fit() for x in model_list]
model_rmse = [rmse(df.y, x.fittedvalues) for x in model_fits]

model_num = 4
print(model_fits[model_num].summary())
print('-------------')
print(model_rmse)
```

```python
from sklearn.model_selection import train_test_split, cross_validate, cross_val_score, KFold
from sklearn import metrics
from sklearn.linear_model import LinearRegression

X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,0:15], df.y) #here's one example of a split
X_train.head()

#Let's do the full cross validate
cv = KFold(n_splits=10, random_state=1, shuffle=True)
cv_scores_raw = [cross_val_score(LinearRegression(),
                                 df.iloc[:,0:1], 
                                 df.y, 
                                 scoring='neg_mean_squared_error',
                                 cv=cv)]

rmse_cv_scores = [np.sqrt(np.abs(np.mean(cross_val_score(LinearRegression(),
                                 df.iloc[:,0:x], 
                                 df.y, 
                                 scoring='neg_mean_squared_error',
                                 cv=cv)))) for x in range(1,16)]
#rmse_cv_scores
```

```python
fig=plt.figure(figsize=(12,8), dpi= 100)

plt.plot(range(1,16), rmse_cv_scores, '-o')
plt.plot(range(1,16), model_rmse[1:16], '-o')
plt.legend(['Cross-Validated MSE', 'Raw MSE'], loc='upper left')
```
